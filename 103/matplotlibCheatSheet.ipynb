```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [10, 15, 7, 12],
    'Value2': [6, 18, 9, 11]
}
df = pd.DataFrame(data)
```

### Bar Charts

#### Horizontal and vertical 

```python
# Plotting using pandas
df.plot(kind='bar', x='Category', y='Value1', legend=False)
plt.title('Bar Graph using Pandas')
plt.ylabel('Values')
plt.show()
```

```python
# Plotting using matplotlib
plt.bar(df['Category'], df['Value1'])
plt.title('Bar Graph using Matplotlib')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()

# using axes
# fig, ax = plt.subplots()
# ax.bar(df['Category'], df['Value1'])
# ax.set_title('Bar Graph using Matplotlib')
# ax.set_xlabel('Category')
# ax.set_ylabel('Values')
# plt.show()
```

```python
# Horizontal 
# Plotting using pandas
df.plot(kind='barh', x='Category', y='Value1', legend=False)
plt.title('Horizontal Bar Chart using Pandas')
plt.xlabel('Values')
plt.show()
```

```python
# using matplotlib 
fig, ax = plt.subplots()
ax.barh(df['Category'], df['Value1'])
ax.set_title('Horizontal Bar Chart using Matplotlib')
ax.set_ylabel('Category')
ax.set_xlabel('Values')
plt.show()
```

#### Stacked

```python
# Using pandas
df.plot(kind='bar', x='Category', stacked=True)
plt.title('Stacked Bar Graph using Pandas')
plt.ylabel('Values')
plt.show()
```

```python
# Plotting using matplotlib
index = range(len(df))

plt.bar(index, df['Value1'], label='Value1')
plt.bar(index, df['Value2'], bottom=df['Value1'], label='Value2')


plt.title('Stacked Bar Graph using Matplotlib')
plt.xlabel('Category')
plt.ylabel('Values')
plt.xticks(index, df['Category'])
plt.legend()
plt.show()


# using axes 
# fig, ax = plt.subplots()
# ax.bar(index, df['Value1'],label='Value1')
# ax.bar(index, df['Value2'],bottom=df['Value1'], label='Value2')

# ax.set_title('Stacked Bar Graph using Matplotlib')
# ax.set_xlabel('Category')
# ax.set_ylabel('Values')
# ax.set_xticks(index)
# ax.set_xticklabels(df['Category'])
# ax.legend()
# plt.show()
```

#### Grouped

```python
# Plotting using pandas 
df.plot(kind='bar', x='Category')
plt.title('Grouped Bar Graph using Pandas')
plt.ylabel('Values')
plt.show()
```

```python
# Plotting using matplotlib
bar_width = 0.35
index = np.arange(len(df))

plt.bar(index - bar_width/2 , df['Value1'], bar_width, label='Value1')
plt.bar(index + bar_width/2, df['Value2'], bar_width, label='Value2')

plt.title('Grouped Bar Graph using Matplotlib')
plt.xlabel('Category')
plt.ylabel('Values')
plt.xticks(index, df['Category'])
plt.legend()
plt.show()



# using axes 
# fig, ax = plt.subplots()
# ax.bar(index - bar_width/2, df['Value1'], bar_width, label='Value1')
# ax.bar(index + bar_width/2, df['Value2'], bar_width, label='Value2')

# ax.set_title('Grouped Bar Graph using Matplotlib')
# ax.set_xlabel('Category')
# ax.set_ylabel('Values')
# ax.set_xticks(index)
# ax.set_xticklabels(df['Category'])
# ax.legend()
# plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = axes.flatten()

ax1.plot(range(10), 'r')
ax2.plot(range(10), 'b')
ax3.plot(range(10), 'g')
ax4.plot(range(10), 'k')
plt.show()
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a figure
plt.figure(figsize=(10, 8))

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.title('Sine')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'b')
plt.title('Cosine')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'g')
plt.title('Tangent')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'k')
plt.title('Exponential')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
