```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [10, 15, 7, 12],
    'Value2': [6, 18, 9, 11]
}
df = pd.DataFrame(data)
```

```python
index = np.arange(len(df))

bar_widht = 0.35 

plt.bar(index - bar_widht / 2, df["Value1"], bar_widht, label="R")
plt.bar(index + bar_widht / 2, df["Value2"], bar_widht, label="D")

plt.xticks(index, df["Category"])
plt.legend(loc = "lower right")
plt.plot()
```
